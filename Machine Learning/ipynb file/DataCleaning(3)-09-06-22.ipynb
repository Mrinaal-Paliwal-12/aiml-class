```python
import pandas as pd
```

```python
ytube = pd.read_csv('D:\\24 - Machine_Learning\\download files\\top-5000-youtube-channels.csv')
ytube
```

```python
ytube.info()
```

```python
ytube['Rank'] = ytube['Rank'].str[0:-2].str.replace(',','').astype('int')
# Here we remove the last 2 character from the data of the rank column
```

```python
ytube
```

```python
ytube.dropna()
# Here we are removing the Nan values
```

```python
ytube['Grade'].unique()
```

```python
mask1 = ytube[ytube['Subscribers'].str.contains('-')].index
ytube.drop(labels = mask1, axis = 0, inplace = True)
```

```python
ytube['Subscribers'] = ytube['Subscribers'].astype('int')
```

```python
mask1
```

```python
ytube.info()
```

```python
mask2 = ytube[ytube['Video Uploads'].str.contains('-')].index
ytube.drop(labels = mask2, axis = 0, inplace = True)
```

```python
ytube['Video Uploads'] = ytube['Video Uploads'].astype('int')
```

```python
mask2
```

```python
ytube.info()
```

```python
ytube['Grade'].unique()
```

```python
channel_map = {'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1}
```

```python
ytube['Grade'] = ytube['Grade'].map(channel_map)
```

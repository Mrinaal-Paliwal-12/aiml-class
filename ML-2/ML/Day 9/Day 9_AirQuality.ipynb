```python
df['Time'].dropna()
df['Time']=pd.to_datetime(df['Time']) 
```

```python
df['Time'].head() # now datatype is datetime
```

```python
df['day']=df["Date"].apply(lambda x: x.day)
df['month']=df["Date"].apply(lambda x: x.month)
df['year']=df["Date"].apply(lambda x: x.year)
```

```python
df.head()
```

```python
df['hour']=df["Time"].apply(lambda x: x.hour)
df['minute']=df["Time"].apply(lambda x: x.minute)
df['second']=df["Time"].apply(lambda x: x.second)

```

```python
df.head()
```

```python
df=df.drop(['Date','Time'],axis=1)
```

```python
print("Number of null values in each column:\n{}".format(df.isnull().sum()))
```

```python
df.info()
```

```python
for i in df.columns:
    df[i]=df[i].fillna(df[i].mean())
```

```python
feat=df.drop(['AH','Unnamed: 15','Unnamed: 16'],axis=1)
val=df['AH'].values
```

```python
feat.head()
```

```python
(train_feat,test_feat,train_classes,test_classes)=train_test_split(feat,val,random_state=40,train_size=0.2)
```

```python
m=DecisionTreeRegressor(max_depth=6).fit(train_feat,train_classes) #for max depth of tree 
#m=DecisionTreeRegressor().fit(train_feat,train_classes)
```

```python
ypred=m.predict(test_feat)
```

```python
from sklearn import metrics

pred=m.predict(test_feat)
print("MSE ",metrics.mean_squared_error(test_classes,ypred))
```

```python
f=feat.values
from sklearn import tree


fig=plt.figure(figsize=(25,20))
_=tree.plot_tree(m,feature_names=f,class_names=val,filled=True)
```

```python
text_representation= tree.export_text(m)
print(text_representation)
```
